```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('datasets/dataset_train.csv')
```

```python
def scatter_plot_(course1, course2):
    houses = ['Gryffindor', 'Slytherin', 'Hufflepuff', 'Ravenclaw']

    fig, ax = plt.subplots(2, 2)
    for i in range(4):
        house = houses[i]
        x = df[course1][df['Hogwarts House'] == house]
        y = df[course2][df['Hogwarts House'] == house]
        ax[int(i % 2)][int(i / 2)].scatter(x,y)
        ax[int(i % 2)][int(i / 2)].legend(house)
    fig.legend([course1, course2])
    
    plt.show()
```

```python

courses = ['Arithmancy',
            'Astronomy',
            'Herbology',
            'Defense Against the Dark Arts',
            'Divination',
            'Muggle Studies',
            'Ancient Runes',
            'History of Magic',
            'Transfiguration',
            'Potions',
            'Care of Magical Creatures',
            'Charms',
            'Flying']
   
    
for each in courses:
    for another in courses:
        if each is not another:
            scatter_plot_(each, another)
```
