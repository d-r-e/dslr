```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
def histogram_(y):
    arr = {}
    for n in y:
        if (str(n) in arr):
            arr[str(n)] += 1
        else:
            arr[str(n)] = 1
    plt.bar(arr.keys(), arr.values())
    return
```

```python
test = [1 ,1, 1, 2, 3, 3 ,3 ,3 , 4, 42, 42, 42, 99]
histogram_(test)
```

```python
df = pd.read_csv('datasets/dataset_train.csv')
courses = ['Arithmancy',
            'Astronomy',
            'Herbology',
            'Defense Against the Dark Arts',
            'Divination',
            'Muggle Studies',
            'Ancient Runes',
            'History of Magic',
            'Transfiguration',
            'Potions',
            'Care of Magical Creatures',
            'Charms',
            'Flying']
for i in range(len(courses)):
    plt.figure(figsize=(10, 4))
    n = df.groupby('Hogwarts House')[courses[i]].plot(kind='hist', alpha=0.60)
    plt.legend()
    plt.title(courses[i])
```
